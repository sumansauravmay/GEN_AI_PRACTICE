```python
# prime sequence number
for num in range(2, 100):
  flag=0
  for val in range(2, num):
    if num % val==0:
      flag=1
      break
  if flag==0:
   print(num, end=" ")
```

```python
num=4
flag=0
for i in range(2, num):
  if num%i==0:
    flag=1

if flag==0:
  print(num, "is prime!")
else:
  print(num, "is not prime!")
```
